```python
import wntr
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Load the water network model
inp_file = '/home/wahab901278/Downloads/Net3.inp'
wn = wntr.network.WaterNetworkModel(inp_file)

# Define contamination scenario
contaminant_node = '125'  # Node where contamination is introduced
contaminant_concentration = 100.0  # Concentration in mg/L
start_time = 2 * 3600  # Start time in seconds
duration = 2 * 3600  # Duration in seconds

# Add contaminant to the network
wn.add_source('ContaminantSource', node_name=contaminant_node, source_type='SETPOINT', quality=contaminant_concentration)

# Run water quality simulation
sim = wntr.sim.EpanetSimulator(wn)
results = sim.run_sim()

# Obtain concentration results
concentration = results.node['quality']

# Sensor locations and observed data (hypothetical example)
sensor_locations = ['120', '121', '122', '123', '124']
observed_data = concentration.loc[start_time + duration:, sensor_locations].max(axis=0)

# Objective function for optimization (minimize error between simulated and observed data)
def objective(x, sensor_locations, observed_data, wn, start_time, duration):
    for i, node in enumerate(sensor_locations):
        wn.add_source(f'Source_{i}', node_name=node, source_type='SETPOINT', quality=x[i])
    sim = wntr.sim.EpanetSimulator(wn)
    results = sim.run_sim()
    sim_data = results.node['quality'].loc[start_time + duration:, sensor_locations].max(axis=0)
    wn.reset_injections()  # Reset sources
    return np.sum((sim_data - observed_data) ** 2)

# Initial guess for source strengths
initial_guess = np.ones(len(sensor_locations)) * 10

# Run optimization
result = minimize(objective, initial_guess, args=(sensor_locations, observed_data, wn, start_time, duration), method='Nelder-Mead')

# Identify potential contamination sources
potential_sources = [sensor_locations[i] for i in range(len(sensor_locations)) if result.x[i] > 0.1]
print("Potential contamination sources based on optimization:")
print(potential_sources)

# Visualize the contamination spread
plt.figure(figsize=(10, 6))
for node in sensor_locations:
    plt.plot(concentration.index, concentration[node], label=f'Node {node}')

plt.title('Contamination Spread in the Network')
plt.xlabel('Time (seconds)')
plt.ylabel('Concentration (mg/L)')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
import wntr
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Load the water network model
inp_file = '/home/wahab901278/Downloads/Net3.inp'
wn = wntr.network.WaterNetworkModel(inp_file)

# Define contamination scenario
contaminant_node = '125'  # Node where contamination is introduced
contaminant_concentration = 100.0  # Concentration in mg/L
start_time = 2 * 3600  # Start time in seconds
duration = 2 * 3600  # Duration in seconds

# Add contaminant to the network
wn.add_source('ContaminantSource', node_name=contaminant_node, source_type='MASS', quality=contaminant_concentration)

# Run water quality simulation
sim = wntr.sim.EpanetSimulator(wn)
results = sim.run_sim()

# Obtain concentration results
concentration = results.node['quality']

# Check which nodes are available in the concentration results
available_nodes = concentration.columns.tolist()

# Define sensor locations and check against available nodes
sensor_locations = ['120', '121', '122', '123', '124']
sensor_locations = [node for node in sensor_locations if node in available_nodes]

# Observed data (hypothetical example)
observed_data = concentration.loc[start_time + duration:, sensor_locations].max(axis=0)

print(observed_data)


# Objective function for optimization (minimize error between simulated and observed data)
def objective(x, sensor_locations, observed_data, wn, start_time, duration):
    for i, node in enumerate(sensor_locations):
        wn.add_source(f'Source_{i}', node_name=node, source_type='MASS', quality=x[i])
    sim = wntr.sim.EpanetSimulator(wn)
    results = sim.run_sim()
    sim_data = results.node['quality'].loc[start_time + duration:, sensor_locations].max(axis=0)
    wn.reset_initial_values()  # Reset sources
    return np.sum((sim_data - observed_data) ** 2)

# Initial guess for source strengths
initial_guess = np.ones(len(sensor_locations)) * 10

# Run optimization
result = minimize(objective, initial_guess, args=(sensor_locations, observed_data, wn, start_time, duration), method='L-BFGS-B')
print(result)
# Identify potential contamination sources
potential_sources = [sensor_locations[i] for i in range(len(sensor_locations)) if result.x[i] > 0.1]
print("Potential contamination sources based on optimization:")
print(potential_sources)

# Visualize the contamination spread
plt.figure(figsize=(10, 6))
for node in sensor_locations:
    plt.plot(concentration.index, concentration[node], label=f'Node {node}')

plt.title('Contamination Spread in the Network')
plt.xlabel('Time (seconds)')
plt.ylabel('Concentration (mg/L)')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
