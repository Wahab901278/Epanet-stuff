## Create a new network

```python
wn=wntr.network.WaterNetworkModel()
```

```python
wntr.graphics.plot_network(wn)
```

```python
#help(wn.add_junction)
#help(wn.add_reservoir)
#help(wn.add_tank)
#help(wn.add_pipe)
#help(wn.add_pump)
help(wn.add_valve)
```

```python
wn.add_junction('J1',base_demand=0.01,elevation=100,coordinates=(1,1))
wn.add_junction('J2',base_demand=0.01,elevation=120,coordinates=(2,2))
wn.add_reservoir('R1',base_head=150,coordinates=(0,0))
wn.add_tank('T1',elevation=130,coordinates=(2,0),init_level=10,min_level=5,max_level=15,diameter=20,min_vol=100)
wn.add_pipe('P1','R1','J1',length=1000,diameter=0.3,roughness=100)
wn.add_pipe('P2','J1','J2',length=1000,diameter=0.3,roughness=100)
wn.add_pump('Pump1','J2','T1',pump_type='POWER',pump_parameter=100)
wn.add_junction('J3',elevation=130,coordinates=(1.5,0))
wn.add_pipe('P3','T1','J3',length=50,diameter=0.3)
wn.add_valve('V1','J3','J1',diameter=0.2,valve_type='PRV',initial_setting=100)
```

```python
wntr.graphics.plot_network(wn,title="New network",node_labels=True,link_labels=True)
```
